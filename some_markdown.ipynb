Hashtables are basically an array that links a key to a specific data value(s), where each key is unique. They are data structure that efficiently implements search/lookup operations. It aims to divide the whole search space into well-defined regions, so we donâ€™t have to search linearly through the entire collection to find an element. 

Hashtable Techniques - 
 - Hashing
 - Hash function
<br>Key values are converted to integer values using <b>hashing technique</b>. And we use the <b>hash function</b> to find the required key-value pair. So, we reduce the search time from linearly to directly searching in the bucket.


<hr>
Ideally for <b>perfect hashing</b> we have only 1 value for each array index. So that we can directly find the index after hashing and get a time complexity of O(1). <br>
